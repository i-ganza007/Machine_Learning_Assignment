Project Plotting - Matplotlib Seaborn, Histograms and Boxplots

```python
import numpy as py
import pandas as pd
import matplotlib.pyplot as plt
```

load the dataset and print 10 rows

```python
fuel_econ = pd.read_csv("fuel_econ.csv")
fuel_econ.head(n=10)
```

```python
plt.figure(figsize=(10,6))
plt.hist(fuel_econ['ghgScore'],bins=10, color='purple', edgecolor='black')
plt.title('Histogram showing the green house gas score of automobiles')
plt.ylabel('Frequency'); plt.xlabel('ghgScore')
plt.grid(True)
plt.show()
```

# Interpretation of the histogram above

The histogram shows the greenhousegas score (ghgScore) of vehicles and how frequent a certain ghgScore was obtained.

On the x-axis, a score of minimum 0 and maximum 10 is showcased, whereby 0 being the sign of the a car that has maximum CO2 emission and 10 being sign of minimum CO2 emission. Preferrably the closer the ghgScore is to 10, the better.

On the y-axis, shown the number of cars that have x ghgScore meaning that the higher the bar goes, the more cars average that specific ghgScore.
